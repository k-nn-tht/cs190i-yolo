```python
import matplotlib.pyplot as plt
```

```python
# Standard model
with open("../standard_model/train_data/lists/losses.txt", "r") as file:
    standard_losses = [float(line.strip()) for line in file]

with open("../standard_model/train_data/lists/mAPs.txt", "r") as file:
    standard_mAP = [float(line.strip()) for line in file]

# AdamW model
with open("../AdamW/train_data/lists/losses.txt", "r") as file:
    AdamW_losses = [float(line.strip()) for line in file]

with open("../AdamW/train_data/lists/mAPs.txt", "r") as file:
    AdamW_mAP = [float(line.strip()) for line in file]

# 9x9 model
with open("../9by9model/train_data/lists/losses.txt", "r") as file:
    nbn_losses = [float(line.strip()) for line in file]

with open("../9by9model/train_data/lists/mAPs.txt", "r") as file:
    nbn_mAP = [float(line.strip()) for line in file]

# 5 bounding boxes model
with open("../5_bounding_boxes/train_data/lists/losses.txt", "r") as file:
    Fivebb_losses = [float(line.strip()) for line in file]

with open("../5_bounding_boxes/train_data/lists/mAPs.txt", "r") as file:
    Fivebb_mAP = [float(line.strip()) for line in file]

# 9x9 with 5 Bounding Boxes model
with open("../9x9&5BB/train_data/lists/losses.txt", "r") as file:
    nbn_5BB_losses = [float(line.strip()) for line in file]

with open("../9x9&5BB/train_data/lists/mAPs.txt", "r") as file:
    nbn_5BB_mAP = [float(line.strip()) for line in file]
```

```python
fig1, axes1 = plt.subplots(1, 2, figsize=(15, 5))

# Plot losses on the first subplot
axes1[0].plot(standard_losses, label="Standard Model")
axes1[0].plot(AdamW_losses, label="AdamW Model")
axes1[0].set_title("Losses")
axes1[0].set_xlabel("Epoch")
axes1[0].set_ylabel("Loss")
axes1[0].legend()
axes1[0].grid(True)

# Plot mAP scores on the second subplot
axes1[1].plot(standard_mAP, label="Standard Model")
axes1[1].plot(AdamW_mAP, label="AdamW Model")
axes1[1].set_title("mAP Scores")
axes1[1].set_xlabel("Epoch")
axes1[1].set_ylabel("mAP")
axes1[1].legend()
axes1[1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

fig2, axes2 = plt.subplots(1, 2, figsize=(15, 5))

# Plot losses on the first subplot
axes2[0].plot(standard_losses, label="Standard Model")
axes2[0].plot(nbn_losses, label="9x9 Model")
axes2[0].plot(Fivebb_losses, label="5 Bounding Boxes Model")
axes2[0].plot(nbn_5BB_losses, label="9x9 & 5BB Model")
axes2[0].set_title("Losses")
axes2[0].set_xlabel("Epoch")
axes2[0].set_ylabel("Loss")
axes2[0].legend()
axes2[0].grid(True)

# Plot mAP scores on the second subplot
axes2[1].plot(standard_mAP, label="Standard Model")
axes2[1].plot(nbn_mAP, label="9x9 Model")
axes2[1].plot(Fivebb_mAP, label="5 Bounding Boxes Model")
axes2[1].plot(nbn_5BB_mAP, label="9x9 & 5BB Model")
axes2[1].set_title("mAP Scores")
axes2[1].set_xlabel("Epoch")
axes2[1].set_ylabel("mAP")
axes2[1].legend()
axes2[1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
